```python
import lungs_finder as lf

image = cv2.imread("Elegiveis/Bacteria/person1000_bacteria_2931.jpeg",0)

# Get both lungs image. It uses HOG as main method,
# but if HOG found nothing it uses HAAR or LBP.
found_lungs = lf.get_lungs(image)
print("Pulmo Encontrado: ", found_lungs.size)

if found_lungs.size > 0:
    plt.figure(figsize=(20,10))
    plt.axis("off")
    plt.imshow(found_lungs)
    plt.title("Suavizao Mediana")


right_lung_hog_rectangle = lf.find_right_lung_hog(image)
left_lung_hog_rectangle = lf.find_left_lung_hog(image)
right_lung_lbp_rectangle = lf.find_right_lung_lbp(image)
left_lung_lbp_rectangle = lf.find_left_lung_lbp(image)
right_lung_haar_rectangle = lf.find_right_lung_haar(image)
left_lung_haar_rectangle = lf.find_left_lung_haar(image)

print("Lado Direito lbp: ", right_lung_lbp_rectangle)
print("Lado Direito haar: ", right_lung_haar_rectangle)
print("Lado Esquerdo lbp: ", left_lung_lbp_rectangle)
print("Lado Esquerdo haar: ", left_lung_haar_rectangle)


if right_lung_lbp_rectangle is not None:
    x, y, width, height = right_lung_lbp_rectangle
    right_image = image[y:y + height, x:x + width]
    plt.figure(figsize=(20,10))
    plt.axis("off")
    plt.title("RIGHT LBP")
    plt.imshow(right_image)

if left_lung_lbp_rectangle is not None:
    x, y, width, height = left_lung_lbp_rectangle
    left_image = image[y:y + height, x:x + width]
    plt.figure(figsize=(20,10))
    plt.axis("off")
    plt.title("LEFT LBP")
    plt.imshow(left_image)

if right_lung_haar_rectangle is not None:
    x, y, width, height = right_lung_haar_rectangle
    right_image = image[y:y + height, x:x + width]
    plt.figure(figsize=(20,10))
    plt.axis("off")
    plt.title("RIGHT HAAR")
    plt.imshow(right_image)

if left_lung_haar_rectangle is not None:
    x, y, width, height = left_lung_haar_rectangle
    left_image = image[y:y + height, x:x + width]
    plt.figure(figsize=(20,10))
    plt.axis("off")
    plt.title("LEFT HAAR")
    plt.imshow(left_image)
```

```python
# Funo para tratar as imagens

def tratamento(nome):
    
    raiox = cv2.imread(nome)
    
    #Sem tratamento
    #Melhor resultado com 300 imagens: val_precision_score: 0.73339 na 2 Epoch
    #imagem_tratada = cv2.cvtColor(raiox, cv2.COLOR_BGR2RGB)
    
    # Escala de cinza de suavizao atravs da mediana
    #Melhor resultado com 300 imagens: val_precision_score: 0.6833 na 3 Epoch
    imagem_cinza = cv2.cvtColor(raiox, cv2.COLOR_BGR2GRAY)
    imagem_tratada = cv2.medianBlur(imagem_cinza, 15)
    
    # Escala de cinza e escurescimento da imagem
    # Melhor resultado com 300 imagens: val_precision_score: 0.6661 na 3 Epoch
    #raiox = cv2.imread(nome, cv2.IMREAD_GRAYSCALE)
    #matriz_brilho = np.ones(raiox.shape, np.uint8) * 100
    #imagem_tratada = cv2.subtract(raiox, matriz_brilho)
    
    
    # Escala de cinza e escurescimento da imagem
    # Melhor resultado com todas as imagens: val_precision_score: 0.4936 
    #raiox = cv2.cvtColor(raiox, cv2.COLOR_BGR2RGB)
    #matriz_brilho = np.ones(raiox.shape, np.uint8) * 100
    #imagem_tratada = cv2.add(raiox, matriz_brilho)
    
    
    raiox = cv2.cvtColor(raiox, cv2.COLOR_BGR2RGB)
    imagem_gray = cv2.cvtColor(raiox, cv2.COLOR_BGR2GRAY)
    imagem_tratada = cv2.Canny(imagem_gray, 1, 100)

    return imagem_tratada
```

```python
#L as imagens elegiveis, submete ao tratamento e salva no diretrio para leitura dos algoritmos

extensao = "*.jpeg"

diretorioOrigem = "Elegiveis/Bacteria/"
diretorioDestino = "raiox/Bacteria/"
for path in glob.glob(diretorioOrigem + extensao):
    
    image = tratamento(path)
    cv2.imwrite(diretorioDestino + path[path.index("\\")+1:], image)


diretorioOrigem = "Elegiveis/Controle/"
diretorioDestino = "raiox/Controle/"
for path in glob.glob(diretorioOrigem + extensao):
    
    image = tratamento(path)
    cv2.imwrite(diretorioDestino + path[path.index("\\")+1:], image)


diretorioOrigem = "Elegiveis/Viral/"
diretorioDestino = "raiox/Viral/"
for path in glob.glob(diretorioOrigem + extensao):
    
    image = tratamento(path)
    cv2.imwrite(diretorioDestino + path[path.index("\\")+1:], image)
```

```python
import cv2
import numpy as np

# Load image as grayscale and crop ROI
image = cv2.imread("Elegiveis/Bacteria/person1_bacteria_1.jpeg",0)

# Calculate mean and STD
mean, STD  = cv2.meanStdDev(image)

# Clip frame to lower and upper STD
offset = 0.2
clipped = np.clip(image, mean - offset*STD, mean + offset*STD).astype(np.uint8)

# Normalize to range
result = cv2.normalize(clipped, clipped, 0, 255, norm_type=cv2.NORM_MINMAX)

#plt.figure(figsize=(20,10))
#plt.axis("off")
#plt.imshow(imagem)
#plt.title("Raio-X")

cv2.imshow('image', image)
cv2.imshow('ROI', ROI)
cv2.imshow('result', result)
cv2.waitKey()
```
